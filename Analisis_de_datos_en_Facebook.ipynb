```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
    # --- Datos de 2024 ---
    'Alcance_2024': [39887, 7258, 8101, 9789, 6845],
    'Interacciones_2024': [95, 144, 320, 254, 178],
    'Visitas_2024': [490, 420, 648, 580, 622],
    'Seguidores_2024': [0, 7, 8, 8, 4],
    # --- Datos de 2025 ---
    'Alcance_2025': [788, 2191, 1829, 3953, 3237],
    'Interacciones_2025': [4, 73, 38, 69, 67],
    'Visitas_2025': [85, 199, 186, 224, 222],
    'Seguidores_2025': [2, 5, 1, 5, 2]
}

df = pd.DataFrame(data)
df.set_index('Mes', inplace=True)

plt.style.use('seaborn-v0_8-whitegrid')
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 14))
fig.suptitle('Análisis Comparativo de Métricas de Facebook (2024 vs. 2025)\nEscuela de Comunicación e Industrias Digitales', fontsize=20, weight='bold')

# --- Gráfico 1: Alcance ---
ax1 = axes[0, 0]
width = 0.40
x = np.arange(len(df.index))

rects1_2024 = ax1.bar(x - width/2, df['Alcance_2024'], width, label='2024', color='#a7c7e7') # Azul claro
rects1_2025 = ax1.bar(x + width/2, df['Alcance_2025'], width, label='2025', color='#c3b1e1') # Morado claro

ax1.set_ylabel('Número de Personas', fontsize=12)
ax1.set_title('Alcance Mensual', fontsize=16, weight='bold')
ax1.set_xticks(x)
ax1.set_xticklabels(df.index, rotation=0, ha='center')
ax1.legend()

ax1.get_yaxis().set_major_formatter(
    plt.FuncFormatter(lambda y, p: format(int(y), ','))
)
ax1.bar_label(rects1_2024, padding=3, fmt='{:,.0f}')
ax1.bar_label(rects1_2025, padding=3, fmt='{:,.0f}')


# --- Gráfico 2: Interacciones con el Contenido ---
ax2 = axes[0, 1]
rects2_2024 = ax2.bar(x - width/2, df['Interacciones_2024'], width, label='2024', color='#a7c7e7') # Azul claro
rects2_2025 = ax2.bar(x + width/2, df['Interacciones_2025'], width, label='2025', color='#c3b1e1') # Morado claro

ax2.set_ylabel('Número de Interacciones', fontsize=12)
ax2.set_title('Interacciones Mensuales', fontsize=16, weight='bold')
ax2.set_xticks(x)
ax2.set_xticklabels(df.index)
ax2.legend()
ax2.bar_label(rects2_2024, padding=3)
ax2.bar_label(rects2_2025, padding=3)


# --- Gráfico 3: Visitas a la Página ---
ax3 = axes[1, 0]
rects3_2024 = ax3.bar(x - width/2, df['Visitas_2024'], width, label='2024', color='#a7c7e7') # Azul claro
rects3_2025 = ax3.bar(x + width/2, df['Visitas_2025'], width, label='2025', color='#c3b1e1') # Morado claro

ax3.set_ylabel('Número de Visitas', fontsize=12)
ax3.set_title('Visitas a la Página', fontsize=16, weight='bold')
ax3.set_xticks(x)
ax3.set_xticklabels(df.index)
ax3.legend()
ax3.bar_label(rects3_2024, padding=3)
ax3.bar_label(rects3_2025, padding=3)


# --- Gráfico 4: Nuevos Seguidores ---
ax4 = axes[1, 1]
rects4_2024 = ax4.bar(x - width/2, df['Seguidores_2024'], width, label='2024', color='#a7c7e7') # Azul claro
rects4_2025 = ax4.bar(x + width/2, df['Seguidores_2025'], width, label='2025', color='#c3b1e1') # Morado claro

ax4.set_ylabel('Número de Seguidores', fontsize=12)
ax4.set_title('Nuevos Seguidores por Mes', fontsize=16, weight='bold')
ax4.set_xticks(x)
ax4.set_xticklabels(df.index)
ax4.legend()
ax4.bar_label(rects4_2024, padding=3)
ax4.bar_label(rects4_2025, padding=3)



fig.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

```

 ANÁLISIS Y RECOMENDACIONES

RESUMEN EJECUTIVO

Al comparar los primeros cinco meses de 2024 con los de 2025, se observa una
disminución general significativa en las métricas clave como el alcance, las
interacciones y las visitas a la página. Enero de 2024 fue un mes
excepcionalmente fuerte, marcando un pico que no se ha replicado en 2025, pero como se menciona en el documento esto se debe a la pauta que se realizó.
Por otra parte, el crecimiento de seguidores también se ha ralentizado considerablemente.
Estos datos sugieren la necesidad de tener una nueva estrategia para
revitalizar la presencia de la marca en Facebook y recuperar el impulso perdido.
Asimismo se puede percibir estas diferencias en estos 5 meses de 2024 y 2025:

- Alcance: el alcance mensual promedio ha caído drásticamente en 2025.
  Enero de 2024 tuvo un alcance total de casi 40,000, mientras que el mes
  más fuerte de 2025 apenas superó las 6,000. Esta caída indica que el
  contenido actual no está siendo priorizado por el algoritmo de Facebook o los ads perjudicaron el algoritmo.

- Interacciones: las interacciones (me gusta, comentarios, etc.) han seguido
  una tendencia similar. Se pasó de picos de más de 300 interacciones
  mensuales en 2024 a promedios de 60-70 en 2025.

- Visitas a la Página: en 2024, era común superar las 500 visitas mensuales.
  En 2025, el promedio se sitúa alrededor de las 200.

- Crecimiento de Seguidores: el ritmo de adquisición se ha estancado, pasando
  de entre 8-20 seguidores por mes en 2024 a cifras mucho menores en 2025.
  
  Algunas sugerencias que se podría realizar en la red social Facebook serían:

1. Priorizar el formato de video (reels): la escuels se debe adaptar al algoritmo creando contenido de video corto que muestre la vida estudiantil, testimonios, etc.

2. Fomentar la comunidad: Como aparece en el informe, esta red social se usa para generar comunidad, se debe crear historias interactivas (encuestas,
   preguntas) y responder a todos los comentarios para aumentar el 'engagement'.

3. Inversión estratégica en publicidad: En dado caso de que no funcione de manera organica al final el año, si debe pensar en designar un presupuesto para promocionar publicaciones clave, como ejemplo mostrando lo que ofrece las ecuela, alianzas o proyectos de estudiantes.

4. Finalmente, colaboraciones y contenido generado por los estudiantes o contactos influyentes: Seguir realizando contenido compartidos con las alianzas estratégicas de la escuela  y destacar el contenido creado por los propios estudiantes para aumentar la autenticidad y el alcance.
